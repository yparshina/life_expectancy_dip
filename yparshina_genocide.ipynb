```python
pop_density = {}
pop_total = {}
pop_urban = {}
pop_rural = {}
pop_urban_perc = {}
pop_rural_perc = {}
birth_rate = {}
death_rate = {}
life_exp_f = {}
life_exp_m = {}
fertility = {}
pop_female = {}
pop_0_14 = {}
pop_15_64 = {}
pop_65_up = {}

import csv
with open("data/Gender_Stats_csv/GenderStat_Data.csv", 'rb') as f:
    social_dev_scv = csv.reader(f)
    for row in social_dev_scv:

        if row[2] == 'Population density (people per sq. km of land area)':
            pop_density[row[0]] = filter(None, row[4:])
            pop_density[row[0]] = [float(i) for i in pop_density[row[0]]]
        
        if row[2] == 'Population, total':
            pop_total[row[0]] = filter(None, row[4:])
            pop_total[row[0]] = [float(i) for i in pop_total[row[0]]]
                
        if row[2] == 'Urban population':
            pop_urban[row[0]] = filter(None, row[4:])
            pop_urban[row[0]] = [float(i) for i in pop_urban[row[0]]]
                
        if row[2] == 'Rural population':
            pop_rural[row[0]] = filter(None, row[4:])
            pop_rural[row[0]] = [float(i) for i in pop_rural[row[0]]]
                
        if row[2] == 'Birth rate, crude (per 1,000 people)':
            birth_rate[row[0]] = filter(None, row[4:])
            birth_rate[row[0]] = [float(i) for i in birth_rate[row[0]]]
            
        if row[2] == 'Death rate, crude (per 1,000 people)':
            death_rate[row[0]] = filter(None, row[4:])
            death_rate[row[0]] = [float(i) for i in death_rate[row[0]]]
                
        if row[2] == 'Life expectancy at birth, female (years)':
            life_exp_f[row[0]] = filter(None, row[4:])
            life_exp_f[row[0]] = [float(i) for i in life_exp_f[row[0]]]
            
        if row[2] == 'Life expectancy at birth, male (years)':
            life_exp_m[row[0]] = filter(None, row[4:])
            life_exp_m[row[0]] = [float(i) for i in life_exp_m[row[0]]]
            
        if row[2] == 'Fertility rate, total (births per woman)':
            fertility[row[0]] = filter(None, row[4:])
            fertility[row[0]] = [float(i) for i in fertility[row[0]]]
        
        if row[2] == 'Population, female (% of total)':
            pop_female[row[0]] = filter(None, row[4:])
            pop_female[row[0]] = [float(i) for i in pop_female[row[0]]]
            
        if row[2] == 'Population ages 0-14 (% of total)':
            pop_0_14[row[0]] = filter(None, row[4:])
            pop_0_14[row[0]] = [float(i) for i in pop_0_14[row[0]]]   
        
        if row[2] == 'Population ages 15-64 (% of total)':
            pop_15_64[row[0]] = filter(None, row[4:])
            pop_15_64[row[0]] = [float(i) for i in pop_15_64[row[0]]]   
            
        if row[2] == 'Population ages 65 and above (% of total)':
            pop_65_up[row[0]] = filter(None, row[4:])
            pop_65_up[row[0]] = [float(i) for i in pop_65_up[row[0]]]   
            
        if row[2] == 'Rural population (% of total population)':
            pop_rural_perc[row[0]] = filter(None, row[4:])
            pop_rural_perc[row[0]] = [float(i) for i in pop_rural_perc[row[0]]]  
        
        if row[2] == 'Urban population (% of total population)':
            pop_urban_perc[row[0]] = filter(None, row[4:])
            pop_urban_perc[row[0]] = [float(i) for i in pop_urban_perc[row[0]]]           
```

# Examining Life Expectancy Dips

```python
year_skip = 5
dip_factor = 3
```

```python
# Dip in men's life expectancy
dip_f = {}
for i in life_exp_f:
    counter_f = 0;
    for j in range(0,len(life_exp_f[i])-year_skip):
        
        if life_exp_f[i][j] > life_exp_f[i][j+year_skip]+dip_factor:
            counter_f+=1
    dip_f[i] = counter_f
    
# Dip in men's life expectancy
dip_m = {}
for i in life_exp_m:
    counter_m = 0;
    for j in range(0,len(life_exp_m[i])-year_skip):
        
        if life_exp_m[i][j] > life_exp_m[i][j+year_skip]+dip_factor:
            counter_m+=1
    dip_m[i] = counter_m
```
