```python
# The dataset comes from the World Bank: http://data.worldbank.org/topic/social-development

# life expectancy dictionaries for women (f) and men (m)
life_exp_f = {}
life_exp_m = {}

import csv
with open("data/social_dev_1.csv", 'rb') as f:
    social_dev_scv = csv.reader(f)
    
    for row in social_dev_scv:
        if 'Life expectancy at birth, female' in row[2]: # select female life expectancy field
            life_exp_f[row[0]] = filter(None, row[4:]) # filter out empty years & add fields sarting with the 4th
            life_exp_f[row[0]] = [float(i) for i in life_exp_f[row[0]]]
                    
        if 'Life expectancy at birth, male' in row[2]:     
            life_exp_m[row[0]] = filter(None, row[4:])
            life_exp_m[row[0]] = [float(i) for i in life_exp_m[row[0]]]
```

```python
# delete empty keys
for i in life_exp_f.keys():
    if not life_exp_f[i]: # delete countries with no entries
        del(life_exp_f[i])
    #if not life_exp_f[i][30:]: # delete countries with 30 or less entries
        #del(life_exp_f[i])
      
for i in life_exp_m.keys():
    if not life_exp_m[i]: # delete countries with no entries
        del(life_exp_m[i])
    #if not life_exp_m[i][30:]: # delete countries with 30 or less entries
        #del(life_exp_m[i])
```

```python
# create min, max, med, and mean dictionaries, for future comparisons
max_life_f = {}
min_life_f = {}
med_life_f = {}
mean_life_f = {}
max_life_m = {}
min_life_m = {}
med_life_m = {}
mean_life_m = {}

for i in life_exp_f:
    max_life_f[i] = int(max(life_exp_f[i]))
    min_life_f[i] = int(min(life_exp_f[i]))
    med_life_f[i] = int(median(life_exp_f[i]))
    mean_life_f[i] = int(mean(life_exp_f[i]))

for i in life_exp_m:
    max_life_m[i] = int(max(life_exp_m[i]))
    min_life_m[i] = int(min(life_exp_m[i]))
    med_life_m[i] = int(median(life_exp_m[i]))
    mean_life_m[i] = int(mean(life_exp_m[i]))
```

```python
#Sorting
sorted( ((v,k) for k,v in mean_life_f.iteritems()), reverse=True)
```

#### Dip in women's life expectancy

```python
year_skip = 6
dip_factor = 6
```

```python
dip_f = {}
for i in life_exp_f:
    counter_f = 0;
    for j in range(0,len(life_exp_f[i])-year_skip):
        
        if life_exp_f[i][j] > life_exp_f[i][j+year_skip]+dip_factor:
            counter_f+=1
    dip_f[i] = counter_f
```

#### Dip in men's life expectancy

```python
dip_m = {}
for i in life_exp_m:
    counter_m = 0;
    for j in range(0,len(life_exp_m[i])-year_skip):
        
        if life_exp_m[i][j] > life_exp_m[i][j+year_skip]+dip_factor:
            counter_m+=1
    dip_m[i] = counter_m
```
